```python
run = wandb.init(project="nyc_airbnb", group="eda", save_code=True)
local_path = wandb.use_artifact("sample.csv:latest").file()
df = pd.read_csv(local_path)
df.head(4)
```

```python
df.info()
```

```python
df.describe()
```

```python
# Set the min price = $10 and max price = $350 as the filter for df
min_price = 10
max_price = 350
idx = df['price'].between(min_price, max_price)
df =df[idx].copy()
```

```python
# Check the filter results
df.describe()
```

```python
#convert last_review column to DateTime
df['last_review'] = pd.to_datetime(df['last_review'])
```

```python
# observe the results of the last_review dtype and the filter (min and max)
df.info()
```
